```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(0, 2 *np.pi)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2 *np.pi)
y = np.sin(x)

plt.ylim([0,1])

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2 *np.pi)
y = np.sin(x)

plt.title("Y=sin(x)(0<y<1)")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.ylim([0,1])

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2 *np.pi)
y = np.sin(x)

plt.title("Y=sin(x)(0<y<1)")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.ylim([0,1])

plt.grid(True)

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2 *np.pi)
y = np.sin(x)

plt.title("Y=sin(x)(0<y<1)")

plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.grid(True)

positions = [0,np.pi/2,np.pi,np.pi*3/2,np.pi*2]
labels = ["0","90","180","270","360"]

plt.xticks(positions, labels)

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2 *np.pi)
y1 = np.sin(x)
y2 = np.cos(x)

plt.title("Y=sin(x)(0<y<1)")

plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.grid(True)

positions = [0,np.pi/2,np.pi,np.pi*3/2,np.pi*2]
labels = ["0","90","180","270","360"]

plt.xticks(positions, labels)

plt.plot(x, y1, color="k")
plt.plot(x, y2, color="b")

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2 *np.pi)
y1 = np.sin(x)
y2 = np.cos(x)

plt.title("Y=sin(x)(0<y<1)")

plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.grid(True)

positions = [0,np.pi/2,np.pi,np.pi*3/2,np.pi*2]
labels = ["0","90","180","270","360"]

plt.xticks(positions, labels)

plt.plot(x, y1, color="k")
plt.plot(x, y2, color="b")

plt.legend(["y1=sin(x)","y2=cos(x)"])
plt.show()
```
