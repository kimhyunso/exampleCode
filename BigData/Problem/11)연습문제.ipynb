# Q2. Income_Category의 카테고리를 다음과 같이 변경하여 newIncome 컬럼에 넣어주세요 ^^ 
# map함수를 활용하셔도, apply를 활용하셔도 상관없습니다 ^^
```
Unknown : N
Less than $40K : a
$40K - $60K : b
$60K - $80K : c
$80K - $120K : d
$120K +’ : e
```

````python
def names(name):
    if name == 'Unkown':
        return 'N'
    elif name == 'Less than $40K':
        return 'a'
    elif name == '$40K - $60K':
        return 'c'
    elif name == '$80K - $120K':
        return 'd'
    else:
        return 'e'
    
bank_df['newIncome'] = bank_df['Income_Category'].apply(lambda x : {x : names(x)})
bank_df
````

# Q3. Customer_Age의 값을 이용하여 나이 구간을 Age_cat이라는 새 컬럼을 만들어주세요.
## (0~9:0 // 10~19:10 // 20~29:20...)

#  각 구간의 빈도수를 출력하라.

````python
def custome_age(age):
    if age <= 9:
        return 0
    elif age <= 19:
        return 10
    elif age <= 29:
        return 20
    elif age <= 39:
        return 30
    elif age <= 49:
        return 40
    elif age <= 59:
        return 50
    elif age <= 69:
        return 60
    elif age <= 79:
        return 70

bank_df['Age_cat'] = bank_df['Customer_Age'].apply(lambda x : custome_age(x))
bank_df['Age_cat'].value_counts()
````

# Q4. Education_Level의 값에서 Graduate이 포함된 경우에는 1 그렇지 않다면 0으로 변경하여
# new_edu_level 컬럼을 정의하고 빈도수를 출력해보세요.

````python
bank_df['new_edu_level'] = bank_df['Education_Level'].str.contains('Graduate').apply(lambda x : 1 if x else 0)
bank_df['new_edu_level'].value_counts()
````

# Q5. Marital_Status 컬럼값이 Married이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그렇지 
# 않은 경우에는 모두 0으로 정의해주세요...

# 그리고 그 결과를 new_status 컬럼에 넣어주세요.

````python
bank_df['newState'] = bank_df.apply(lambda x : 1 if x.Marital_Status == 'Married' and x.Card_Category == 'Platinum' else 0, axis=1)

bank_df['newState'].value_counts()
````
