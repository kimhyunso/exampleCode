# Q7 거주인주, 근무인구와 방문인구, 총 유동인구로만 구성된 df를 만드세요

```python
city_df = jeju_df.loc[:, ['거주인구', '근무인구', '방문인구', '총 유동인구']]
city_df
```

# Q8 위에서 생성된 df에 행을 추가해주세요
## 추가 행값은 min값으로 해주세요

```python
min_거주인구 = jeju_df.거주인구.mean()
min_근무인구 = jeju_df.근무인구.mean()
min_방문인구 = jeju_df.방문인구.mean()
min_총_유동인구 = jeju_df['총 유동인구'].mean()

city_df.loc[city_df.shape[0]]= [min_거주인구,min_근무인구,min_방문인구, min_총_유동인구]
city_df
```
