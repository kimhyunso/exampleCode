```python
import numpy as np
import pandas as pd

chip_df = pd.read_csv('./chipotle/chipotle.tsv', sep='\t')
```

# Q1. 데이터의 정보를 확인해주세요

```python
chip_df.info()
```

# Q2. quantity의 값이 3이상인 데이터들을 호출하고
## 전체 수가 얼마인지 출력해주세요.

```python
chip_df[chip_df.quantity >= 3].shape[0]
```

# Q3. item_price의 컬럼의 달러표시 문자를 없애고, float64bit 타입으로
## 저장하여 new_price를 컬럼으로 하여 넣어주세요.

```python
chip_df['new_price'] = chip_df.item_price.apply(lambda x : x[1:-1]).astype('float64')
chip_df
```

# Q4. new_price의 컬럼값이 5이하인 데이터프레임을 추출하고 그 수를 구해주세요.

```python
chip_df[chip_df.new_price <= 5].shape[0]
```

# Q5 new_price의 컬럼값이 9이하이고 item_name값이 Chicken Salad Bowl인 데이터
## 프레임을 추출해주세요.

```python
chip_df[(chip_df.new_price <= 9) & (chip_df.item_name == 'Chicken Salad Bowl')].reset_index()
```

# Q6. 데이터 프레임 new_price의 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화하라

```python
chip_df.sort_values(by='new_price').reset_index()
```

# Q7. 데이터 프레임의 item_name컬럼 값 중에서 Steak가 포함된 경우의 데이터의 수를 구해주세요.

```python
chip_df[chip_df['item_name'].str.contains('Steak')].item_name.value_counts()
```

# Q8. 데이터 프레임 new_price 컬럼값에 따라 내림차순으로 정리하고 index를 초기화해주세요.

```python
chip_df.sort_values(by='new_price', ascending=False).reset_index()
```

# Q9. 데이터 프레임 new_price 값이 new_price값의 평균값 이상을 가지는 데이터들을 인덱싱해주세요.

```python
chip_df[chip_df.new_price >= chip_df.new_price.mean()]
```

# Q10. 데이터 프레임 item_name에서 '6 Pack Soft Drink'의 값을 'pepsi'로 변형해주세요.

```python
# 1. lambda를 활용한 식
chip_df['item_name'] = chip_df.item_name.apply(lambda x : 'pepsi' if x == '6 Pack Soft Drink' else x)
chip_df
```

```python
# 2. loc를 활용한 식 loc[row :: 조건식, col] 
chip_df.loc[chip_df.item_name == '6 Pack Soft Drink', 'item_name'] = 'pepsi'
```

# Q11. 전체 결측값(Missing Value)에 대한 갯수를 구하고, 그 값을 'N'으로 대체해주세요.

```python
chip_df.isna().sum()
chip_df = chip_df.fillna('N')
chip_df
```

# Q12. 데이터 프레임에서 choice_description 컬럼 값에서 Black이 들어가지 않는 경우를 인덱싱 해주세요.

```python
chip_df[~chip_df.choice_description.str.contains('Black')]
```

# Q13. 데이터 프레임에서 item_name 값의 단어의 수가 15개 이상인 데이터를 인덱싱 해주세요.

```python
# 1. lambda를 사용한다.
chip_df[chip_df.item_name.apply(lambda x : len(x) >= 15)]
```

```python
# 2. condition indexing
chip_df[chip_df.item_name.str.len() >= 15]
```
