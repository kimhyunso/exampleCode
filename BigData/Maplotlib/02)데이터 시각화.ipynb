# 여러 그래프

```python
# y1 = 2x^2
# y2 = 5x +30
# y3 = 4x^2 + 10
import numpy as np
x = np.arange(-4.5, 5.0, 0.5)
y1 = 2*x**2
y2 = 5*x + 30
y3 = 4*x**2 + 10
```

```python
## 여러 개의 함수 그리기
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
## 2번째 방법
plt.plot(x, y1, x, y2, x, y3)
plt.show()
```

```python
## 3번째 방법
plt.plot(x, y1)
plt.figure()

plt.plot(x, y2)
plt.figure()
```
