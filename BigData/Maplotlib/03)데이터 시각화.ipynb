# 선그래프

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = [10, 14, 19, 20, 25]
plt.plot(data1)
plt.show()
```

```python
# X데이터와 y = 2x^2
import numpy as np
x = np.arange(-4.5, 5, 0.5)
y = 2*x**2
plt.plot(x, y)
plt.show()
```

# 여러 그래프

```python
# y1 = 2x^2
# y2 = 5x + 30
# y3 = 4x^2 + 10


x = np.arange(-4.5, 5, 0.5)
y1 = 2 * x**2
y2 = 5 * x + 30
y3 = 4 * x**2 + 10

# 여러 개의 함수 그래프 그리기 - 1
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()

# 여러 개의 함수 그래프 그리기 - 2
plt.plot(x, y1, x, y2, x, y3)
plt.show()
```

```python
# 새로운 창에 그래프를 생성해보자.
plt.plot(x, y1) # 처음 그림

# 새 창을 추가
plt.figure() # 새로운 창 추가
plt.plot(x, y2) # 새로운 창에 직선이 그려진다.
plt.show()
```

```python
# 1번 데이터쌍(x, y1) 2번 데이터쌍(x, y2)를 지정된 번호로 그래프 창에 입력해보자


# y1 = x^2 - 2
# y2 = 20 * cos(x)^2

x = np.arange(-5, 5, 0.1)

y1 = x**2 - 2
y2 = 20 * np.cos(x)**2
y3 = 20 * np.sin(x)**2
y4 = 20 * np.tan(x)**2
plt.figure(1) # 1번 그래프 창에 넣어주세요
plt.plot(x, y1)
plt.show()
# ---------------------
plt.figure(2)
plt.plot(x, y2)

plt.figure(2)
plt.plot(x, y3)
plt.show()
# ---------------------
plt.figure(3)
plt.plot(x, y3)
plt.show()
# ---------------------
plt.figure(4)
plt.plot(x, y4)
plt.show()
```
