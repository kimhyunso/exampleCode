```python
import matplotlib.pyplot as plt
import seaborn
import numpy as np
import pandas as pd

plt.rcParams['font.family'] = 'Malgun Gothic' # 윈도우즈의 '맑은 고딕'으로 설정
plt.rcParams['axes.unicode_minus'] = False
```

```python
fruit = ['사과', '바나나', '딸기', '오렌지', '포도']
result = [7, 6, 3, 2, 2]
```

```python
plt.pie(result)
plt.show()
```

```python
# 생성한 원이 타원이므로 원형으로 변환한다. 타원 -> 원형
plt.figure(figsize=(5, 5))
plt.pie(result, labels=fruit, autopct='%.1f%%')
plt.show()
```

- default는 각 부채꼴은 x축 기준 각도를 0도를 시작으로 반시계방향으로 각 부채꼴 부분이 그려진다. 
- 만약 x축 기준 각도 90도에서 시작하여 시계방향으로 설정하여 각 부채꼴 부분을 표시하려면 파이그래프 옵션 지정으로 변형이 가능하다.

```python
plt.figure(figsize=(5, 5))
# counterclock = False :: 시계방향결정 False:시계방향, True:반시계방향, startangle=90 90도부터시작

plt.pie(result, labels=fruit, autopct='%.1f%%', radius=0.8, startangle=90, counterclock=False)
plt.show()
```

```python
explode_value = (0.1, 0, 0, 0, 0,)
plt.figure(figsize=(5, 5))

plt.pie(result, labels=fruit, autopct='%.1f%%', radius=0.8, startangle=90, counterclock=False, shadow=True, explode=explode_value)
# img 저장하기
plt.savefig('./save_test0.png')

plt.show()
```
