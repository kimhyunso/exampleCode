Seaborn은 Matplotlib의 기능과 스타일을 확장한 파이썬 시각화 도구의 고급버전이다. 비교적 단순하므로 초보자도 쉽게 따라 할 수 있다.

```python
# import library
import seaborn as sns

# 타이타닉 데이터셋 로딩
titan_df = sns.load_dataset('titanic')

```

# 회귀직선이 있는 산점도

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 스타일을 지정 - DarkGrid
sns.set_style('darkgrid')
```

```python
# 그래프 객체 생성 - 한 out에 그래프를 살펴보도록
fig = plt.figure(figsize=(15, 5)) # 15, 5 // 16, 8

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# ax1 = fig.add_subplot(1, 3, 2)
# ax1 = fig.add_subplot(1, 3, 2)
# ax1 = fig.add_subplot(1, 3, 2)


# 그래프 그리기 - 선형회귀선 표시 (fit_reg = True)
sns.regplot(x='age', y='fare', data=titan_df, ax=ax1)

# 그래프 그리기 - 선형회귀선 미표시 (fit_reg = Flase)
sns.regplot(x='age', y='fare', data=titan_df, ax=ax2, fit_reg=False)

plt.show()
```

# 히스토그램 및 밀도함수

```python
fig = plt.figure(figsize=(15, 5)) # 15, 5 // 16, 8

ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

# 기본값
sns.distplot(titan_df['fare'], ax=ax1)

# hist = False
sns.distplot(titan_df['fare'], ax=ax2, hist=False)


# kde = False
sns.distplot(titan_df['fare'], ax=ax3, kde=False) # Kernel Density


# 차트 제목 표시
ax1.set_title('titan fare-hist/kde')
ax2.set_title('titan fare-hist')
ax3.set_title('titan fare-kde')
plt.show()
```

# 히트맵(Heat-map)

```python
table = titan_df.pivot_table(index=['sex'], columns=['class'], aggfunc='size')
```

```python
# 히트맵 그리기

# fmt = d ::  데시멀 // f :: float
# sns.heatmap(table, annot=False, fmt='d', linewidths=0.5)
# sns.heatmap(table, annot=True, fmt='.2f', linewidths=10)
# linewidths 간격조정
sns.heatmap(table, annot=True, fmt='d', linewidths=10) # heat-map을 쓰실 때,

plt.show()
```

# 박스플롯 / 바이올린 플롯

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 데이터 로드
titan_df = sns.load_dataset('titanic')

# set_style theme
sns.set_style('whitegrid')

# 그래프 객체 생성
fig = plt.figure(figsize=(15, 5))


ax = list()

for i in range(1, 5):
    ax.append(fig.add_subplot(2, 2, i))
    
# 박스 플롯 - 기본값
sns.boxplot(x='alive', y='age', data=titan_df, ax=ax[0])

# 박스 플롯 - hue 변수 추가
sns.boxplot(x='alive', y='age', data=titan_df, ax=ax[1], hue='sex')


# 바이올린 플롯 - 기본값
sns.violinplot(x='alive', y='age', data=titan_df, ax=ax[2])


# 바이올린 플롯 - hue 변수 추가
sns.violinplot(x='alive', y='age', data=titan_df, hue='sex', ax=ax[3])

plt.show()

```
