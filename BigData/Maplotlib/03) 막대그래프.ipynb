|회원ID|운동시작전|운동시작후|
|---|---|---|

```python
members_id = ['m_01', 'm_02', 'm_03', 'm_04']
before_work = [27, 35, 40, 33]
after_work = [30, 38, 42, 37]
```

```python
# plt.bar

import matplotlib.pyplot as plt
import numpy as np

n_data = len(members_id) # 4명
# Numpy 배열 (0, 1, 2, 3)
index_1 = np.arange(n_data)

# bar(x, y)에서 x=index, y=before_work
plt.bar(index_1, before_work)
plt.show()

plt.bar(index_1, before_work, tick_label=members_id)
plt.show()
```

```python
# 가로 막대 그래프를 생성
colors = ['r', 'g', 'b', 'm']

plt.barh(index_1, before_work, color=colors)
plt.show()
```

```python
# 하나의 장에서 두 개의 그래프를 그리자
barwidth = 0.4
plt.bar(index_1, before_work, color='c', align='edge', width=barwidth, label='before')
plt.bar(index_1+barwidth, after_work, color='m', align='edge', width=barwidth, label='after')
plt.legend() # 범례
plt.xlabel('회원 ID')
plt.ylabel('근육량')
plt.title('운동 전후의 근육량 변화율표')
plt.show()
```
