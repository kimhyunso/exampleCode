```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], 'ro') # 리스트, 튜플, np.array
plt.axis([0, 7, 0, 15])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.arange(0, 5, 0.2)
# 빨간 대쉬, 녹색 사각형, 파랑 삼각형
plt.plot(data, data, 'r--', data, data**2, 'gs', data, data**3, 'b^')
plt.show()

```
