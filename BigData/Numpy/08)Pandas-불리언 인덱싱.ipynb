```python
import pandas as pd
titanic_df = pd.read_csv('./titanic/train.csv')
titanic_df
```

```python
# 2. 60세 초과인 사람들 -> reset_index 까지...
titanic_df[titanic_df['Age'] > 60].reset_index().head(3)
```

```python
# 2. 성별이 여성인 승객 조건을 만족하는 '이름', '나이' 컬럼을 추출해보자
titanic_df[titanic_df['Sex'] == 'female'][['Name', 'Age']] 
```

```python
# 객실등급이 1등급인 사람들을 추출해보자
titanic_df[titanic_df['Pclass'] < 2]['Pclass'].value_counts()
```

# Q1. 위의 세 조건을 결합해보자
1. 60세 이상
2. 여성
3. 선실등급 1등급

```python
titanic_df[(titanic_df['Sex'] == 'female') & (titanic_df['Age'] > 60) & (titanic_df['Pclass'] < 2)]
```

# Q2. Q1번을 부정문으로 써보세요.. - 문제를 위한 문제..

```python
titanic_df[~((titanic_df['Sex'] == 'female') & (titanic_df['Age'] > 60) & (titanic_df['Pclass'] < 2))]
```
