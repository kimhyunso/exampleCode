```python
# 결측치 - Missing Value를 확인한다
titan_df.isna().sum()
```

```python
# 결측치 대치(imputation)
# 결측치 대체 - 1) 최대값, 2) 최소값, 3) 대표값(최빈, 중앙, 평균), 4) 해당컬럼삭제
age_mean = titan_df.Age.mean()

titan_df.Age.fillna(age_mean, inplace=True) # inplace 파라미터를 기억하자
titan_df.Cabin.fillna('N', inplace=True)

titan_df.Embarked.fillna('N', inplace=True)
```
