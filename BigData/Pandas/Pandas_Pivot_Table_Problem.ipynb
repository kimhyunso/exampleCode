```python
import numpy as np
import pandas as pd
import seaborn as sns
titan_df = pd.read_csv('../scv_files/titanic/train.csv')
```

```python
titan_df.pivot_table(index='Sex', values='Survived')
```

```python
titan_df.pivot_table(columns='Sex', values='Survived')
```

```python
titan_df.pivot_table(index=['Sex', 'Pclass'], values='Survived')
```

```python
titan_df.pivot_table(index='Sex', columns='Pclass', values='Survived')
```

```python
titan_df.pivot_table(index='Sex', columns='Pclass', values='Survived', aggfunc=['mean', 'sum'])
```

|day|Thur|Fri|Sat|Sun|
