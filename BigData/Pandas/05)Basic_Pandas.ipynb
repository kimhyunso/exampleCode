## 행과 열을 추가 삭제

```python
# 열을 추가하자
titan_df['Age_0'] = 0

# 파생변수 생성 - 연구자의 논리적 주관에 따라 데이터 정보에 맞게 생성된 변수
titan_df['Fam_no'] = titan_df.SibSp + titan_df.Parch + 1

titan_df['Age_by_100'] = titan_df.Age * 100
titan_df
```

```python
titan_df.drop(['Age_0', 'Age_by_100'], axis=1, inplace=True)
```

```python
titan_col = titan_df.columns
titan_col[-2]
```

```python
# 열을 삭제하자 drop() 함수
# 1. 원본에서 컬럼을 지우는 방법
# titan_df.drop('Ageaa', axis=1, inplace=True)

# 2. 원본은 그대로 컬럼을 지우는 방법
titan_df = titan_df.drop(['Age_0', 'Age_by_100'], axis=1)
```
